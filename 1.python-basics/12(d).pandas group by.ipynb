```python
# no of movie that start from a
def alp_a(group):
    return (group["Series_Title"].str.startswith("A").sum())
```

```python
movies.groupby("Genre").apply(alp_a)
```

```python
# Find ranking ogf eacxh movie in the movie accordinding to IMDB ranking
def rank_move(group):
    group["IMDB_Rating"] = group["IMDB_Rating"].rank(ascending=False)
    return group

genre.apply(rank_move)
```

```python
# normalized rating on basis og group
def normalized(group):
    group["norm_rating"] = (group["IMDB_Rating"]-group["IMDB_Rating"].min())/(group["IMDB_Rating"].max() - group["IMDB_Rating"].min())
    return group
genre.apply(normalized)
    
```

```python
ipl = pd.read_csv("deliveries.csv")
```

```python
deliveries.columns
```

```python
# find the top 10 bats man in terms of batsmaruns
ipl.groupby("batsman")["batsman_runs"].sum().sort_values(ascending = False).head()
```

```python
# find the batsman who have nmax no of sixes
six = ipl[ipl["batsman_runs"] == 6]
six.groupby('batsman')["batsman"].count().sort_values(ascending=False).head()
```

```python
# find batsman whos have most number of 4's & 6's in last 5 overs
temp_df=ipl[ipl['over'] >15]
temp_df=temp_df[(temp_df["batsman_runs"] == 4) | (temp_df["batsman_runs"] == 6)]
temp_df.groupby("batsman")["batsman"].count().sort_values(ascending=False).head(1).index[0]
```

```python
# find "V Kohli" record against all terms
temp_df = ipl[ipl["batsman"] == "V Kohli"]
temp_df.groupby("bowling_team")["batsman_runs"].sum().reset_index()
# no of overs, 
```

```python
# Create a function that returns height score
temp_df = ipl[ipl["batsman"] == "V Kohli"]
temp_df.groupby("match_id")["batsman_runs"].sum().sort_values(ascending=False).head(1).values[0]
```

```python
def highest(batsman):
  temp_df = ipl[ipl['batsman'] == batsman]
  return temp_df.groupby('match_id')['batsman_runs'].sum().sort_values(ascending=False).head(1).values[0]
```

```python
highest("V Kohli")
```

```python
highest("DA Warner")
```
