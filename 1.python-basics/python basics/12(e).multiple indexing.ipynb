### multi index series

```python
# can we have multiple index? Let's try
index_val = [('cse',2019),('cse',2020),('cse',2021),('cse',2022),('ece',2019),('ece',2020),('ece',2021),('ece',2022)]
a = pd.Series([1,2,3,4,5,6,7,8],index=index_val)
a
```

```python
# The problem?
a['cse']
```

```python
# The solution -> multiindex series(also known as Hierarchical Indexing)
# multiple index levels within a single index
```

```python
# how to create multiindex object
# 1. pd.MultiIndex.from_tuples()
index_val = [('cse',2019),('cse',2020),('cse',2021),('cse',2022),('ece',2019),('ece',2020),('ece',2021),('ece',2022)]
multiindex = pd.MultiIndex.from_tuples(index_val)
multiindex.levels[1]
# 2. pd.MultiIndex.from_product()
pd.MultiIndex.from_product([['cse','ece'],[2019,2020,2021,2022]])
```

```python
# creating a series with multiindex object
s = pd.Series([1,2,3,4,5,6,7,8],index=multiindex)
s
```

```python
# how to fetch items from such a series
s['cse']
```

```python
s[('ece',2020)]
```

```python
# unstack can convert 2d to dataframe
```

```python
temp=s.unstack()
```

```python
# stack can convert 2d to dataframe
```

```python
temp.stack()
```

# Long vs wide data format

<img src="https://qph.cf2.quoracdn.net/main-qimg-dd0dabfa8abd51156906518c0e04f684-pjlq">

- Wide format is where we have a single rows for every data point with mutiple columns to hold the calues of various attributes.

- long data format is where, for each data point we have many ros as the number of attributes and each rowscontains the value of a particular attributes for a given data point.

```python
# melt wide data to long data and viz.
```

```python
pd.DataFrame({"cse":[120]}).melt()
```

```python
pd.DataFrame({"cse":[120],'ece':[100],'mech':[12]}).melt()
```

```python
pd.DataFrame({"cse":[120],'ece':[100],'mech':[12]}).melt(var_name='branch',value_name='no_of_students')
```

```python
temp=pd.DataFrame(
{
    'branch':['cse','ece','mech'],
    '2020':[100,150,60],
    '2021':[120,130,80],
    '2022':[150,140,70]
})
```

```python
temp.melt()
```

```python
temp.melt(id_vars=["branch"],var_name='year',value_name="students")
```
