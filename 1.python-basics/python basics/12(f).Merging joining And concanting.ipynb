```python
import pandas as pd
import numpy as np
```

```python
courses = pd.read_csv('courses.csv')
students = pd.read_csv('students.csv')
nov = pd.read_csv('reg-month1.csv')
dec = pd.read_csv('reg-month2.csv')

matches = pd.read_csv('matches.csv')
delivery = pd.read_csv('deliveries.csv')
```

## pd.concat

```python
# concating two month data (vertically stacking)
regs = pd.concat([nov,dec],ignore_index=True)
```

```python
# alternative
nov.append(dec, ignore_index=True)
```

```python
## Multi-index dataframe
multi= pd.concat([nov,dec],keys=["Nov","Dec"])
```

```python
multi.loc["Nov"] # iloc for integer index & loc for label index
```

```python
multi.loc[("Dec",4)]
```

```python
## inner joining side by side
pd.concat([nov,dec], axis=1) # nan for not prsent in less size data
```

```python
## Merge (innser join)
students.merge(regs,how="inner",on="student_id")
```

```python
# left join
courses.merge(regs, how="left", on="course_id")
```

```python
# right jion
temp= pd.DataFrame({
    "student_id": [26,27,28],
    "name":["Nitish","Ankit","Rahul"],
    "partner":[28,26,17]
})
pd.concat([students,temp],ignore_index=True)
```

```python
# right join
students.merge(regs,how="right",on="student_id")
```

```python
# left join
students.merge(regs,how="left",on="student_id")
```

```python
# FULL-OUTER JION
students.merge(regs,how="outer",on="student_id")
```

```python
# 1 find the total revrenue generated
total=regs.merge(courses,how="inner",on="course_id")["price"]
total.sum()
```

```python
# month by month revenue
tem=pd.concat([nov,dec],keys=["Nov","Dec"]).reset_index()
tem.merge(courses,on="course_id").groupby(["level_0"])["price"].sum()
```

```python
# print registration table (name course, price)
regs.merge(students,on="student_id").merge(courses,on="course_id")[["name","course_name","price"]]
```

```python
# bar chart for rvenue& course
regs.merge(courses,on="course_id").groupby("course_name")["price"].sum().plot(kind="bar")
```

```python
#  find students who enrolled both in nov and dec
common_stu = np.intersect1d(nov["student_id"],dec["student_id"])
common_stu
```

```python
students[students["student_id"].isin(common_stu)]
```

```python
# find course that got no enrolledment
course_id_list = np.setdiff1d(courses["course_id"],regs["course_id"])
courses[courses["course_id"].isin(course_id_list)]
```

```python
# find student who did not enrolled inb any course
# find course that got no enrolledment
student_id_list = np.setdiff1d(students["student_id"],regs["student_id"])
students[students["student_id"].isin(student_id_list)]

```

```python
# print student name & partner name for all enrolled student
# self join
students.merge(students,how="inner",left_on="partner", right_on="student_id")[["name_x","name_y"]]
```

```python
# find top3 students who did most enrolled
regs.merge(students, on='student_id')["name"].value_counts().head(3)
# if names are not repeating
```

```python
regs.merge(students, on='student_id').groupby(["student_id","name"])["name"].count().sort_values(ascending=False).head(3)
```

```python
## STUDENTS HOW PAY MOST
regs.merge(students,on="student_id").merge(courses,on="course_id").groupby(["student_id","name"])["price"].sum().sort_values(ascending=False).head()
```

```python
# Alternate syntax for merge
pd.merge(students,regs,how="inner",on='student_id')
```

```python
## top three stadium with highest sixes per matches
```

```python
temp=delivery.merge(matches,left_on="match_id",right_on="id")
```

```python
six_df = temp[temp["batsman_runs"]==6]
```

```python
# stadium -> sixes
num_sixes=six_df.groupby("venue")["venue"].count()
```

```python
num_matches = matches["venue"].value_counts()
```

```python
(num_sixes/num_matches).sort_values(ascending=False).head(3)
```

```python
### FIND ORANGE CAP EVERY STADIUM
temp.groupby(["season", "batsman"])["batsman_runs"].sum().reset_index().sort_values('batsman_runs',ascending=False).drop_duplicates(subset=["season"],keep="first").sort_values('season')
```

```python
### FIND ORANGE CAP EVERY STADIUM
temp.groupby(["season", "batsman"])["batsman_runs"].sum().reset_index().sort_values('batsman_runs',ascending=False)
```
