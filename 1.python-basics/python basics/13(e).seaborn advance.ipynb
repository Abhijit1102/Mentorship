### Figure level function -> `catplot`

```python
sns.scatterplot(data=tips, x='total_bill',y='tip')
```

```python
# strip plot
# axes level function
sns.stripplot(data=tips,x='day',y='total_bill')
```

```python
# using catplot
# figure level function
sns.catplot(data=tips, x='day',y='total_bill',kind='strip')
```

```python
# jitter, hue
# using catplot
# figure level function
sns.catplot(data=tips, x='day',y='total_bill',kind='strip', jitter=0.2, hue="sex")
```

```python
# swarmplot
sns.catplot(data=tips, x="day", y="total_bill", kind="swarm")
```

```python
# swarmplot
sns.swarmplot(data=tips, x="day", y="total_bill",hue="sex")
```

### Boxplot

A boxplot is a standardized way of displaying the distribution of data based on a five number summary (“minimum”, first quartile [Q1], median, third quartile [Q3] and “maximum”). It can tell you about your outliers and what their values are. Boxplots can also tell you if your data is symmetrical, how tightly your data is grouped and if and how your data is skewed.

image.png

```python
# Box plot
sns.boxplot(data=tips,x='day',y='total_bill')
```

```python
# using catplot
# Box plot
sns.catplot(data=tips,x='day',y='total_bill',kind="box", hue="sex")
```

```python
# Box plot with hue
sns.boxplot(data=tips,x='day',y='total_bill',hue="sex")
```

```python
# single boxployt
# Box plot
sns.boxplot(data=tips,y='total_bill')
```

```python
# Violinplot = Boxplot + kdeplot
```

```python
# Box plot
sns.violinplot(data=tips,x='day',y='total_bill')
```

```python
# Box plot
sns.catplot(data=tips,x="day",y='total_bill',kind="violin",hue="sex",split=True)
```

```python
# barplot
sns.barplot(data=tips,x="sex",y="total_bill")
```

- When there arte multiple observstion in each catagories, it also used bootstrapping to compute a confidence interval arround estimate, which is ploted using error bar

```python
# barplot
sns.barplot(data=tips,x="sex",y="total_bill",errorbar=('ci',0))
```

```python
# barplot
sns.catplot(data=tips,x="sex",y="total_bill",kind="bar",hue="smoker",estimator="median")
```

```python
# barplot
sns.catplot(data=tips,x="sex",y="total_bill",kind="bar",hue="smoker",estimator=np.var)
```

```python
# point plot
# barplot
sns.pointplot(data=tips,x="sex",y="total_bill",hue="smoker",estimator=np.mean)
```

```python
# point plot
# barplot
sns.pointplot(data=tips,x="sex",y="total_bill")
```

# countplot 
- Aspecial case for bar plot is used when you want to show the number of observation in each catagory rather than computing a statistics for a second variable,This is simillar to histigram over a catagorical, rather than qualiative variable.

# Regression plot
- regplot
- lmplot
   - In the simplest invocation, both function draw a scatterplot of two vatables,x and y and also fit regression model y ~ x and plot the resulting regression line and a 95% confidence interval for that regression

```python
sns.scatterplot(data=tips, x="total_bill",y='tip')
```

```python
sns.regplot(data=tips, x="total_bill",y='tip')
```

```python
sns.lmplot(data=tips, x="total_bill",y='tip',hue="sex",x_estimator=np.mean)
```

# residplot

```python
sns.residplot(data=tips, x="total_bill", y="tip", 
    robust=True, color="#990201")
```

# multiplot

- FacetGrid: figure level plot( rel, dis, cat) can used facet

```python
sns.catplot(data=tips,x="sex", y="total_bill",kind="violin",col="day",row="time")
```

# Alternative of above with FacetGrid 

```python
g=sns.FacetGrid(data=tips,col="day",row="time")
g.map(sns.violinplot,"sex","total_bill")
g.add_legend()
```

# pairplot

```python
iris.head()
```

```python
sns.pairplot(iris,kind="kde",hue="species")
```

# Pair Grid

```python
g = sns.PairGrid(iris,hue="species")
g.map(sns.scatterplot)
```

```python
# map_diag -L map_offdiag
g = sns.PairGrid(iris,hue="species")
g.map_diag(sns.histplot)
g.map_offdiag(sns.scatterplot)
```

```python
# map_diag -L map_offdiag
g = sns.PairGrid(iris,hue="species")
g.map_diag(sns.boxplot)
g.map_offdiag(sns.histplot)
```

```python
# map_diag -L map_offdiag
g = sns.PairGrid(iris,hue="species")
g.map_diag(sns.histplot)
g.map_upper(sns.kdeplot)
g.map_lower(sns.scatterplot)
```

```python
# vars
# map_diag -L map_offdiag
g = sns.PairGrid(iris,hue="species", vars=["sepal_width","petal_width"])
g.map_diag(sns.histplot)
g.map_upper(sns.kdeplot)
g.map_lower(sns.scatterplot)
```
