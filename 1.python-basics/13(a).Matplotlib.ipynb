```python
# from a pandas DataFrame
batsman = pd.read_csv("sharma-kohli.csv")
batsman
```

```python
plt.plot(batsman["index"],batsman["V Kohli"])
```

```python
# plotting multiple plots
plt.plot(batsman["index"],batsman["V Kohli"])
plt.plot(batsman["index"],batsman["RG Sharma"])
```

```python
# labels title
# plotting multiple plots
plt.plot(batsman["index"],batsman["V Kohli"],color="green",linestyle="dotted", label="V Kohli")
plt.plot(batsman["index"],batsman["RG Sharma"],linestyle="dashdot", label="RG Sharma")

plt.title("R.sharma Vs V.kohli")
plt.xlabel("seasons(in years)")
plt.ylabel("Runs")

plt.legend(loc="best")
```

```python
# plotting a simple function
price = [48000,54000,57000,49000,47000,45000,100000]
year = [2015,2016,2017,2018,2019,2020,2021]

plt.plot(year, price) #catagorical vs numerical
plt.ylim(0,75000) # triming vistualization
plt.xlim(2016,2020) # limmiting axis
```

```python
# grid
# labels title
# plotting multiple plots
plt.plot(batsman["index"],batsman["V Kohli"],color="green",linestyle="dotted", label="V Kohli")
plt.plot(batsman["index"],batsman["RG Sharma"],linestyle="dashdot", label="RG Sharma")

plt.title("R.sharma Vs V.kohli")
plt.xlabel("seasons(in years)")
plt.ylabel("Runs")

plt.legend(loc="best")
plt.grid()

plt.show()
```

```python
# plt.scatter
x = np.linspace(-10,10,50)
y = 10*x + 3 + np.random.randint(0,100,50)

plt.scatter(x,y)
```

```python
# in pandas as pd
df=pd.read_csv("batter.csv")
df = df.head(50)
```

# barplot
- Bivariate
- numerical vs catagorical
- use case - Aggreagate

```python
chd =[10,20,40,55,60,70]
color=["green","yellow","black","red","pink","white"]

plt.barh(color,chd,)
#if number of catagories has increase use horizontal bar
```

```python
#color and label
df = pd.read_csv("batsman_season_record.csv")
```

```python
df
```

```python
plt.bar(df["batsman"],df["2015"])
plt.bar(df["batsman"],df["2016"])
plt.bar(df["batsman"],df["2017"])
```

```python
chd
```

```python
# xticks
# multiple bar plot
plt.bar(np.arange(df.shape[0])-0.2,df["2015"], width=0.2)
plt.bar(np.arange(df.shape[0]),df["2016"], width=0.2)
plt.bar(np.arange(df.shape[0])+0.2,df["2017"], width=0.2)

plt.title("srason wise runs")
plt.xlabel("Players")
plt.ylabel("Runs")

plt.xticks(np.arange(df.shape[0]), df["batsman"])

plt.show()
```
