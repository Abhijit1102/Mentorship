```python
sns.scatterplot(data=tips,x="total_bill",y="tip",hue="sex",style="time",size="size")
plt.show()
```

# kdeplot (Kernel Density  Estimation) 
- rather than using discrete bins, a kde plot smoths the observation with gaussian kernel, producing a continious density estimate

```python
sns.kdeplot(data=tips,x="total_bill")
```

```python
# figure level function
sns.displot(data=tips, x='total_bill', kind="kde", hue="sex", fill=True)
```

```python
# hue -> fill
```

# rug plot
- plot marginal distribution by drawing ticks along the x and y axis
- This function is intended to complement other plots by showing the location of individual  obsevation in an unobstructive ways

```python
sns.kdeplot(data=tips, x='total_bill',fill=True)
sns.rugplot(data=tips, x='total_bill')
```

# Matrixplot

- Heatmap
- Clustering

```python
# Haet map

# plot rectangular data as a colour-encoded matrix
```

```python
gap.head()
```

```python
temp_df =gap.pivot(index="country",columns="year",values="lifeExp")
```

```python
temp_df.head()
```

```python
# axis level function
plt.figure(figsize=(15,15))
sns.heatmap(temp_df,cmap="summer")
```

```python
# annot
temp = gap[gap["continent"] == "Europe"].pivot(index="country",columns="year",values="lifeExp")
temp.head()
```

```python
# annot, linewith
plt.figure(figsize=(15,15))
sns.heatmap(temp, annot=True, linewidth=0.5, cmap="summer")
```
